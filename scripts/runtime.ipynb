```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../output.csv', names=['alt', 'az', 'prediction', 'energy', 'datetime'], parse_dates=['datetime'])
df = df.set_index(df.datetime).drop(columns=['datetime'])

df.shape
```

```python
aggregated_rate = df.prediction.groupby(pd.Grouper(freq='1s')).count()

plt.scatter(aggregated_rate.index, aggregated_rate, color='gray', alpha=0.5, s=2)
plt.plot(aggregated_rate.rolling(window='10s').mean())

plt.xlim([aggregated_rate.index.min(), aggregated_rate.index.max()])
```
